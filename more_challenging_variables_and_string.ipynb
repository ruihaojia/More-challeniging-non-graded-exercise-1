REUIREMENTS:
1. Use the string from the 'final_text' variable to produce an output that's shown under 'OUTPUT'. 
2. You are allowed to use only the methods of split() and strip().
3. You will also need to use 'for' loop and 'if-elif' statement.


OUTPUT:

Name: Charles Dickens
Date of Birth: February 7, 1812
Birthplace: Portsmouth, England
Most Famous For: a writer of the Victorian era
Married To: Catherine Hogarth
Year of Marriage: 1836
Number of Children: ten
Famous Works: "A Christmas Carol", "Oliver Twist", "David Copperfield", and "Great Expectations"
Other Talent: doing dramatic public readings of my work
Date of Death: June 9, 1870
Place of Death: Gad's Hill Place, Kent, England

```python
final_text = """Hello. My name is Charles Dickens.
I am an English novelist.
I was born on February 7, 1812, in Portsmouth, England.
I am most famous as a writer of the Victorian era.
I married Catherine Hogarth in 1836. We have ten children.
Some of my most famous works include "A Christmas Carol", "Oliver Twist", "David Copperfield", and "Great Expectations".
Besides being a writer, my other talent was doing dramatic public readings of my work.
I died on June 9, 1870, at my home in Gad's Hill Place, Kent, England."""

# Split the final_text into lines
lines = final_text.split('\n')

# Initialize variables to hold the extracted information
name = ""
dob = ""
birthplace = ""
famous_for = ""
married_to = ""
year_of_marriage = ""
number_of_children = ""
famous_works = ""
other_talent = ""
date_of_death = ""
place_of_death = ""

# Process each line to extract the information
for line in lines:
    line = line.strip()
    if line.startswith("Hello. My name is"):
        name = line.replace("Hello. My name is", "").strip() + " Dickens"
    elif line.startswith("I was born on"):
        parts = line.split("I was born on")[1].split(", in")
        dob = parts[0].strip()
        birthplace = parts[1].strip()
    elif line.startswith("I am most famous as"):
        famous_for = line.replace("I am most famous as", "").strip()
    elif line.startswith("I married"):
        parts = line.split("I married")[1].split(" in")
        married_to = parts[0].strip()
        year_of_marriage = parts[1].split(".")[0].strip()
    elif "children" in line:
        number_of_children = line.split("We have")[1].split("children")[0].strip()
    elif line.startswith("Some of my most famous works include"):
        famous_works = line.split("Some of my most famous works include")[1].strip()
    elif line.startswith("Besides being a writer"):
        other_talent = line.split("Besides being a writer,")[1].strip()
    elif line.startswith("I died on"):
        parts = line.split("I died on")[1].split(" at my home in")
        date_of_death = parts[0].strip()
        place_of_death = parts[1].strip()

# Print the formatted output
print(f"Name: {name}")
print(f"Date of Birth: {dob}")
print(f"Birthplace: {birthplace}")
print(f"Most Famous For: {famous_for}")
print(f"Married To: {married_to}")
print(f"Year of Marriage: {year_of_marriage}")
print(f"Number of Children: {number_of_children}")
print(f"Famous Works: {famous_works}")
print(f"Other Talent: {other_talent}")
print(f"Date of Death: {date_of_death}")
print(f"Place of Death: {place_of_death}")


```
